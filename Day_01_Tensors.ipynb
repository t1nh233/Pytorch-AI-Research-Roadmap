- T·∫°o ma tr·∫≠n `A` shape `(3, 3)` v√† vector `b` shape `(1, 3)`.
- Th·ª±c hi·ªán `A + b`. Gi·∫£i th√≠ch t·∫°i sao c·ªông ƒë∆∞·ª£c d√π kh√°c k√≠ch th∆∞·ªõc?
- Th·ª±c hi·ªán nh√¢n ma tr·∫≠n: `A @ A` (ho·∫∑c `torch.matmul`).

```python
#Tao ma tran A va vector b

A = torch.randn(3, 3)
b = torch.randn(1, 3)

print("Ma tran A:\n", A)
print("Vector b:\n", b)
```

```python
#Thuc hien A + b

sum_Ab = A + b
print("Ket qua cua phep tinh: A + b\n", sum_Ab)

#Giai thich vi broadcasting duoc tren shape cua A va cua b
```

```python
#Phep nhan ma tran A * (A + b)

mulA_sumAb = A @ (A + b)
print("Ket qua cua phep tinh: A * (A + b)\n", mulA_sumAb)

#hoac co the su dung matmul nhu sau

mulA_sumAb_ = torch.matmul(A, A + b)
print("Ket qua cua phep tinh (su dung matmul): A * (A + b)\n", mulA_sumAb_)

#kiem tra xem co giong nhau khong

print("Giong nhau khong: ", torch.allclose(mulA_sumAb, mulA_sumAb_))
```

- T·∫°o tensor `x` shape `(2, 3, 4)`.
- D√πng `.view()` ho·∫∑c `.reshape()` ƒë·ªÉ ƒë·ªïi th√†nh `(2, 12)`.
- D√πng `.permute(0, 2, 1)` ƒë·ªÉ ƒë·ªïi th√†nh `(2, 4, 3)`. In `.shape` ra ƒë·ªÉ ki·ªÉm tra.

```python
#Tao tensor thu voi shape (2, 3, 4)

tensor_thu = torch.randn(2, 3, 4)
print("Tensor vua duoc tao voi shape (2, 3, 4): \n", tensor_thu)

#Dung view va reshape de doi shape thanh (2, 12)

tensor_thu_view = tensor_thu.view(2, 12)
tensor_thu_reshape = tensor_thu.reshape(2, 12)

print("Tensor sau khi dung view: \n", tensor_thu_view)
print("Tensor sau khi dung reshape: \n", tensor_thu_reshape)

#Dung permute(0, 2, 1) de doi thanh (2, 4, 3)

tensor_thu_permute = tensor_thu.permute(0, 2, 1)
print("Tensor ban dau sau khi permute(0, 2, 1): \n", tensor_thu_permute)
```

B√†i t·∫≠p √¥n t·∫≠p cu·ªëi bu·ªïi

```python
##Khoi tao du lieu tho

batch_images = torch.randn((10, 3, 64, 64))

print(f"Shape g·ªëc: {batch_images.shape}")
print(f"ƒê·ªãa ch·ªâ g·ªëc: {batch_images.data_ptr()}")
```

```python
##Trich xuat & luu tru

flat_images = batch_images.view(10, -1)
print(f"Shape sau khi flatten: {flat_images.shape}")
print(f"ƒê·ªãa ch·ªâ sau khi flatten: {flat_images.data_ptr()}")


##Khi thay doi gia tri dau tien cua flat_images xem batch_images co thay doi khong

print(f"Gia tri batch_images goc (ban dau): {batch_images[0, 0, 0, 0]}")
flat_images[0, 0] = 999.0
print(f"Gia tri batch_images sau khi thay doi: {batch_images[0, 0, 0, 0]}")
```

```python
##Xoay truc va tuong tac voi stride

permuted_images = torch.permute(batch_images, (0, 2, 3, 1))
print(f"Shape sau khi permute: {permuted_images.shape}")
print(f"ƒê·ªãa ch·ªâ sau khi permute: {permuted_images.data_ptr()}")

print(f"Stride cua batch_images: {batch_images.stride()}")
print(f"Stride cua permuted_images: {permuted_images.stride()}")

##Kiem tra xem permuted_images co contiguous

print(f"permuted_images co contiguous: {permuted_images.is_contiguous()}")

##Khong lien tiep (contiguous) thi cach giai quyet

contiguous_images = permuted_images.contiguous()

##Kiem tra lai sau khi giai quyet

print(f"contiguous_images co contiguous: {contiguous_images.is_contiguous()}")
print(f"Stride cua contiguous_images: {contiguous_images.stride()}")
```

```python
##Su ket noi giua numpy va GPU

np_images = batch_images.numpy()

print(f"Mean (NumPy): {np.mean(np_images):.4f}")

# ƒê∆∞a l√™n GPU (N·∫øu c√≥)
if torch.cuda.is_available():
    print("ƒêang ƒë·∫©y l√™n GPU...")
    batch_gpu = batch_images.to('cuda')

    # Th·ª≠ convert ng∆∞·ª£c v·ªÅ NumPy (S·∫Ω l·ªói)
    try:
        tmp = batch_gpu.numpy()
    except TypeError as e:
        print(f"L·ªñI B·∫ÆT ƒê∆Ø·ª¢C: {e}")
        print("-> Gi·∫£i ph√°p: Ph·∫£i d√πng .cpu().numpy()")
        tmp = batch_gpu.cpu().numpy()
        print("ƒê√£ fix th√†nh c√¥ng!")
else:
    print("M√°y kh√¥ng c√≥ GPU ƒë·ªÉ test ph·∫ßn n√†y.")
```

```python
##Broadcasting tensor

brightness = torch.tensor([0.1, 0.2, 0.3]).view(3, 1, 1)
print(f"Brightness shape: {brightness.shape}")

result = brightness + batch_images
print(f"Result shape: {result.shape}")
```
